# Pandas prepisovanie údajov

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/home/dell/DATA_ZALOHA/Data_Science/Data stes Pandas/fotbal_prestupy_2000_2019.csv')
```

### Vytvorenie nového stĺpca 
- s náhodnými číslami

```python
# Vytvorenie stĺpcu s náhodnými číslami (Pozor bude to float lebo sú to desatinné čísla)
col = pd.Series(np.random.random (4700)) # Definovaná premenná col
display (col)
```

### Do DataFrame priradený stĺpec

```python
# Priradený k DataFrame vygenerovaný nový stĺpec ktorý som pomenoval
df['kvalita_trenera'] = col # Pomenovaný stĺpec (kvalita_trenera)
display (df)
```

### Sčítanie dvoch stĺpcov do jedného stĺpca

```python
# v DataFrame definovaný nový stĺpec
# sčítanie dvoch stĺpcov (Přestupová částka, kvalita_trenera )
# Sčítavané stĺpce ale ostávajú v DataFrame
df['hodnota'] = df['Přestupová částka'] + df['kvalita_trenera']
print (df)
```

### Vytvorenie nového stĺpca, ktorý informuje o inom stĺpci
- bool operácia 

```python
# Vytvorený nový stĺpec z názvom Záložník
# Definujeme stĺpec v ktorom budeme prehľadávať
# Hľadaný výraz (Central Midfilder)
df['Záložník'] = df.Pozice == 'Central Midfield'
print (df)
```
